Your task is to add two columns. One for the latitude and one for the longitude.
Note: with the structure of type series
you cannot use any methods, only you
can use apply function.

```python
add = pandas.read_csv("supermarkets.csv") 
add['Locations']= (add['Address']+","+add["City"]+","+add['State']+','+add["Country"]).apply(nom.geocode)
# Adds Latitude column
add['Latitude'] = add['Locations'].apply(lambda x: x.latitude if x else None)
# Adds Latitude column
add['Longitude'] = add['Locations'].apply(lambda x: x.longitude if x else None)

add
```
